Numpy

```python
import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Print the array
print(arr)

# Array operations
print(arr * 2)  # Multiply each element by 2
print(arr.sum())  # Calculate the sum of elements
print(arr.mean())  # Calculate the mean of elements
```

Pandas

```python
import pandas as pd

# Create a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 22]}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)

# Access data by column name
print(df['Name'])

# Add a new column
df['City'] = ['New York', 'Los Angeles', 'Chicago']

# Filter data
filtered_df = df[df['Age'] > 25]
```

Matplotlib

```python
import matplotlib.pyplot as plt

# Generate some data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]

# Create a line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Line Plot')

# Show the plot
plt.show()
```

Seaborn 

```python
import seaborn as sns

# Load built-in dataset (optional, replace with your data)
iris = sns.load_dataset('iris')

# Create a scatter plot with color-coded species
sns.scatterplot(
    x='sepal_length', y='sepal_width', hue='species', data=iris
)

# Show the plot
plt.show()
```
