```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score
```

```python
# Load the Iris dataset
iris_df = pd.read_csv('iris.csv')

iris_df
```

```python
# Split the dataset into features and target variable
X = iris_df.drop('Species', axis=1)
y = iris_df['Species']
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a logistic regression model
logreg = LogisticRegression()

# Create Recursive Feature Elimination (RFE) object
rfe = RFE(logreg, n_features_to_select=1)

# Fit RFE
rfe = rfe.fit(X_train, y_train)
```

```python

# Print the selected features
print("Selected Features:")
for i in range(len(rfe.support_)):
    if rfe.support_[i]:
        print(X.columns[i])

# Transforming the data
X_train_rfe = rfe.transform(X_train)
X_test_rfe = rfe.transform(X_test)

# Fit the logistic regression model using the selected features
logreg.fit(X_train_rfe, y_train)

# Make predictions
y_pred = logreg.predict(X_test_rfe)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron

# Load the Iris dataset
iris = load_iris()

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25)

# Create the perceptron
perceptron = Perceptron()

# Fit the perceptron to the training data
perceptron.fit(X_train, y_train)

# Make predictions on the test data
y_pred = perceptron.predict(X_test)

# Calculate the accuracy
accuracy = np.mean(y_pred == y_test)

# Print the accuracy
print("Accuracy:", accuracy)
```
